### 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions. [4 Marks]

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
from scipy.stats import binom
```

```python
n=10
p=0.05
k=np.arange(0,11)
k
```

```python
A = stats.binom.pmf(n=n,p=p,k=k)
A
```

#### 2.A. Probability that none of the items are defective? [1 Mark]

```python
print(A[0]) # A[0] is Probability of 0 items defective.
print("Probability that none of the items are defective is %.2f" %(A[0]*100),"%")
```

#### 2.B. Probability that exactly one of the items is defective? [1 Mark]

```python
print(A[1]) # A[1] is Probability of 1 items defective.
print("Probability that exactly one of the items is defective is %.2f" %(A[1]*100),"%")
```

#### 2.C. Probability that two or fewer of the items are defective? [1 Mark]

```python
k2 = 2 #k2 is the number of trials which are successes.
```

```python
B = stats.binom.cdf(n=n,p=p,k=k2)
print(B)
print("Probability that two or fewer of the items are defective is %.2f" %(B*100),"%")
```

#### 2.D. Probability that three or more of the items are defective ? [1 Mark]

```python
# We can calculate the probability from left side only hence we subract the value from "1"
D = 1-stats.binom.cdf(n=n,p=p,k=k2) 
print(D)
print("Probability that three or more of the items are defective is %.2f" %(D*100),"%")
```

### 3. A car salesman sells on an average 3 cars per week. [3 Marks]

```python
# Here Average = 3, We have only average value with time hence this is a poission distribution.
```

```python
avg_sell = 3 #(Mean)
n = np.arange(0,16)
n
```

```python
poisson = stats.poisson.cdf(n,avg_sell)
poisson
```

#### 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]

```python
# P = probability that a car salesman sells some cars. Which is total selling excluded from none selling.
P = 1-poisson[0]
print(P)
print("Probability that a car salesman sells some cars is %.2f" %(P*100),"%")
```

#### 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]

```python
# Let P1 be the probability that salesman sells less than 5 cars
P1 = poisson[4] # P(X < 5) = Probability upto 4.
# Let P2 be the probability that salesman sells more than 1 car
P2 = poisson[1] # P(X => 2) = Probability upto 1.
P3 = P1 - P2 # Prob. that the salesman will sell between 2 and 5 cars
print(P3)
print('Probability that salesman sells between 2 and 5 cars is %.2f'%(P3*100),"%")
```

#### 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]

```python
poisson = stats.poisson.cdf(n,avg_sell)
plt.plot(n,poisson, 'o-')
plt.title('Poisson distribution')
plt.xlabel('Number of Cars sold per week')
plt.ylabel('Cumulative Prob. of Cars sold per week')
plt.show()
```

#### G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants: [2 Mark]

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# Operating winners = A
# Operating contestants = B

# H0 : A = B (There is no significant difference in proportion between winners & contestants of Operating companies)
# H1 : A != B (There is significant difference in proportion between winners & contestants of Operating companies)

significance = 0.025 #Alpha

# our samples - 5.64% is winner percentage and 94.36% is Contestant percentge

sample_success_a, sample_size_a = (18, 319)
sample_success_b, sample_size_b = (301, 319)

# check our sample against H0 for H1 != H0
successes = np.array([sample_success_a, sample_success_b])
samples = np.array([sample_size_a, sample_size_b])

stat, p_value = proportions_ztest(count=successes, nobs=samples)
```

#### H. Test for significance and conclusion [1 Mark]

```python
# Report
print('z_stat: %0.5f, p_value: %0.5f' % (stat, p_value))
if p_value > significance:
   print ("Fail to reject the null hypothesis - There is no significant difference in proportion between winners & contestants of Operating companies")
else:
   print ("Reject the null hypothesis - There is significant difference in proportion between winners & contestants of Operating companies")
 
```
